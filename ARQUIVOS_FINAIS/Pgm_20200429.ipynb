```python
# Arquivos Spharm

dfOEndo = pd.read_csv('OctantEndoSPHARM_01_20200414.txt', header=None)  #, header=None)

dfOEndo.head(15)
```

```python
# Junção das colunas
dfP1 = dfP.copy()
for l in range(0, 11):
    for n in range(1, 9):
        
        if (l < 10):
            nmc = 'OctEndoL0' + str(l) + '0' + str(n)
        else:
            nmc = 'OctEndoL' + str(l) + '0' + str(n)
        
        dfP1[nmc] = 0.00

# Cria as colunas do Octant Endo de L0 a L10
linha = 0
ct = 0
for l in range(len(dfOEndo)):
    id = dfOEndo.iloc[l, 0]
    
    # Montagem o nome da coluna
    nmL = dfOEndo.iloc[l, 1]
    if (len(nmL) == 12):
        nmc = 'OctEndoL0' + nmL[11]
    else:
        nmc = 'OctEndoL' + nmL[11:13]
    
    # Inclusão das 8 colunas
    for c in range(2, 10):
        nmcl = nmc + '0' + str(c - 1)
        dfP1.loc[linha, nmcl] = dfOEndo.iloc[l, c]
    
        if (dfP1.iloc[linha, 0] != dfOEndo.iloc[l, 0]):
            print('**** Erro: ', dfP1.iloc[linha, 0], dfOEndo.iloc[l, 0])   

    if (ct > 9):
        ct = 0
        linha += 1
    else:
        ct += 1
        
dfP1.head()   
```

```python
# Arquivos Total Spharm

dfTSEndo = pd.read_csv('TotalEndoSPHARM_01_20200414.txt', header=None)  #, header=None)

dfTSEndo.head(15)
```

```python
# Junção das colunas
dfP4 = dfP3.copy()

# Construção das colunas
for n in range(0, 11):     
    if (n < 10):
        nmc = 'TotalEndoSpharmL0' + str(n) 
    else:
        nmc = 'TotalEndoSpharmL' + str(n) 
    dfP4[nmc] = 0.00

# Atualiza as colunas novas com a somatória de cada L
linha = 0
ct = 0
for l in range(len(dfTSEndo)):
    id = dfTSEndo.iloc[l, 0]
    
    # Montagem o nome da coluna
    nmL = dfTSEndo.iloc[l, 1]
    if (len(nmL) == 12):
        nmc = 'TotalEndoSpharmL0' + nmL[11] 
    else:
        nmc = 'TotalEndoSpharmL' + nmL[11:13] 
        
    # Somatório das colunas de cada L
    soma = 0.00
    tc = len(dfTSEndo.columns)
    for c in range(2, tc):
        if (pd.isna(dfTSEndo.iloc[l, c])):
            continue
        else:
            soma += dfTSEndo.iloc[l, c]
    
    # Inclusão da coluna com a somatória
    dfP4.loc[linha, nmc] = soma
    
    # Verificação de consistência entre os conjuntos de linhas das duas bases
    if (ct > 9):
        if (dfP4.iloc[linha, 0] != dfTSEndo.iloc[l, 0]):
            print('**** Erro: ', dfP4.iloc[linha, 0], dfTSEndo.iloc[l, 0])   

        ct = 0
        linha += 1
    else:
        ct += 1
        
dfP4.head()
```
